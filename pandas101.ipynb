1. เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean() [['Detached House','Condo','Twin house','Townhome']]
```

2. พื้นที่ (area_total_min_wa และ area_usable_min) และจำนวนห้องนอน
ของทรัพย์สินที่แพงที่สุดแต่ละชนิด

```python
unit.groupby(['propertytype_name_en']).max() [['price_min','area_total_min_wa','area_usable_min','count_room_bed']]
```

```python
maxprice = unit.groupby(['propertytype_name_en'])[['price_min','area_total_min_wa','area_usable_min','count_room_bed']].max().reset_index()
maxprice
```

```python
for row in maxprice.iterrows():
  print(row[1]['propertytype_name_en'],row[1]['price_min'])
  print(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']])
```

```python
(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']]) #ดูค่าสุดท้ายของลูป
```

3. จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด

```python
unit.groupby(['propertytype_name_en']).min() [['price_min','count_parking','count_room_bath']]
```

```python
minprice = unit.groupby(['propertytype_name_en'])[['price_min','count_parking','count_room_bath']].min().reset_index()
minprice
```

```python
for row in minprice.iterrows():
  print(row[1]['propertytype_name_en'],row[1]['price_min'])
  print(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['count_parking','count_room_bath']])
```
